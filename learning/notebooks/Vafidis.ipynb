- 数量
    - head-direction cells(头朝向细胞) 数量: $N_{\mathrm{HD}} = 60$, 形成 ring. 
    - head-rotation cells(头旋转细胞) 数量: $N_{\mathrm{HR}} = 60$
        - left HR cells(左旋细胞) 数量: $N_{\mathrm{LHR}}=N_{\mathrm{HR}}/2 = 30$
        - right HR cells(右旋细胞) 数量: $N_{\mathrm{RHR}}=N_{\mathrm{HR}}/2 = 30$

- 放电率
    - $t$ 时刻 HD firing rate(头朝向细胞放电率) $\vec{r}_{\mathrm{HD}}(t) \in \mathbb{R}^{N_{\mathrm{HD}}}$
        - 第 $i$ 个 HD cell 的 firing rate: $r_{\mathrm{HD},i}(t)$
    - $t$ 时刻 HR firing rate(头旋转细胞放电率) $\vec{r}_{\mathrm{HR}}(t) \in \mathbb{R}^{N_{\mathrm{HR}}}$
        - 第 $j$ 个 HR cell 的 firing rate: $r_{\mathrm{HR},j}(t)$

- 权重连接矩阵
    - HD $\to$ HD: **rec**urrent matrix, $\mathbf{W}_{\mathrm{HD}\to\mathrm{HD}}\in\mathbb{R}^{N_{\mathrm{HD}}\times N_{\mathrm{HD}}}$, 可塑
    - HR $\to$ HD: $\mathbf{W}_{\mathrm{HR}\to\mathrm{HD}} \in \mathbb{R}^{N_{\mathrm{HD}}\times N_{\mathrm{HR}}}$, 可塑
    - HD $\to$ HR: $\mathbf{W}_{\mathrm{HD}\to\mathrm{HR}} \in \mathbb{R}^{N_{\mathrm{HR}}\times N_{\mathrm{HD}}}$, 处理为 fixed

真实头朝向 $\theta(t)\in [0,2\pi]$, 真实角速度 $v(t) \in \mathbb{R}$

第 $i$ 个 HD cell 的 preferred direction(偏好方向) $\theta_{\mathrm{HD},i} \in [0,2\pi]$, 均匀分布在 ring 上. 

- Visual input(视觉输入) 矢量 $\vec{I}_{\mathrm{vis}\to\mathrm{HD}} \in \mathbb{R}^{N_{\mathrm{HD}}}$: periodic Gaussian-like bump
    - 视觉输入被传递到 HD cell 的 axon-proximal compartment(轴突近端区)
    - 第 $i$ 个分量是围绕真实头朝向 $\theta(t)$ 的 local bump: $\begin{aligned}I_{\mathrm{vis}\to\mathrm{HD},i}(t) = A_{\mathrm{vis}}\exp{\left[-\frac{1}{2\sigma_{\mathrm{vis}}^{2}}\sin^{2}\left(\frac{\theta_{\mathrm{HD},i}-\theta(t)}{2}\right)\right]}\end{aligned}$
        - $A_{\mathrm{vis}} \in \mathbb{R}$: 视觉 bump 幅值
        - $\sigma_{\mathrm{vis}} \in \mathbb{R}^{+}$: 视觉 bump 宽度
        - $I_{\mathrm{vis},0} \in \mathbb{R}^{-}$: 抑制性 baseline

- 角速度输入矢量 $\vec{I}_{\mathrm{vel}\to\mathrm{HR}}(t) \in \mathbb{R}^{N_{\mathrm{HR}}}$
    - (角)速度输入被传递到 HR cell
    - 第 $j$ 个分量为 $I_{\mathrm{vel}\to\mathrm{HR},j}(t) = q_{\mathrm{HR},j}k_{\mathrm{vel}}v(t)$
        - $k_{\mathrm{vel}}\in\mathbb{R}^{+}$: 速度到输入电流的比例系数
        - $\begin{aligned}q_{\mathrm{HR},j} = \begin{cases} -1, j\leq N_{\mathrm{HR}}/2 \\  +1, j > N_{\mathrm{HR}}/2\end{cases}\end{aligned}$, 用于区分 left HR 和 right HR. 











HD distal input current(HD 远端输入电流) 矢量 $\vec{I}_{\mathrm{HD},d}(t) \in \mathbb{R}^{N_{\mathrm{HD}}}$

- 动力学: $\begin{aligned}\tau_{s}\frac{\mathrm{d}\vec{I}_{\mathrm{HD},d}}{\mathrm{d}t} = -\vec{I}_{\mathrm{HD},d} + \mathbf{W}_{\mathrm{HD}\to\mathrm{HD}}\vec{r}_{\mathrm{HD}} + \mathbf{W}_{\mathrm{HR}\to\mathrm{HD}}\vec{r}_{\mathrm{HR}} + I_{\mathrm{inh},\mathrm{HD}}\vec{1}_{N_{\mathrm{HD}}}\end{aligned}$
    - $\tau_{s}$: synaptic time constant(突触时间常数)
    - $I_{\mathrm{inh},\mathrm{HD}}\in\mathbb{R}$: HD 集群的 global inhibition(全局抑制), $\vec{1}_{N_{\mathrm{HD}}}\in\mathbb{R}^{N_{\mathrm{HD}}} = \begin{bmatrix} 1 \\ 1 \\ \vdots \\ 1 \end{bmatrix}$








- HD cell 被处理为 2-compartment neuron(双室神经元)
    - HD axon-distal voltage(远端树突电压) 矢量 $\vec{V}_{\mathrm{HD},d}(t) \in \mathbb{R}^{N_{\mathrm{HD}}}$
        - toy model 可令 $\vec{V}_{\mathrm{HD},d}(t) = \vec{I}_{\mathrm{HD},d}(t)$
    - HD distal predition(远端预测) 矢量 $\vec{V}_{\mathrm{HD},ss}(t) = p_{d\to a}\vec{V}_{\mathrm{HD},d}(t)$
        - $\begin{aligned}p_{d\to a} = \frac{g_{d\to a}}{g_{d\to a} + g_{L}}\in(0,1)\end{aligned}$
            - $g_{d\to a}$: distal-to-proximal compling conductance(远端-近端耦合电导)
            - $g_{L}$: leak conductance(漏电导)
    - HD axon-proximal voltage(近端树突电压) 矢量 $\vec{V}_{\mathrm{HD},a}(t) \in \mathbb{R}^{N_{\mathrm{HD}}}$
        - toy model 可写作 $\vec{V}_{\mathrm{HD},a}(t) = \vec{V}_{\mathrm{HD},ss}(t) + \vec{I}_{\mathrm{vis}\to\mathrm{HD}}(t)$, 近端电压等于远端预测+视觉 teacher 信号
    - sigmoid activation function(激活函数) $\begin{aligned}f(x) = \frac{f_{\max}}{1 + \exp{[-\beta(x-x_{1/2})]}}\end{aligned}$
        - $f_{\max}$: 最大 firing rate
        - $\beta$: 斜率
        - $x_{1/2}$: 半最大输入
    - HD firing rate(头朝向细胞放电率) 矢量 $\vec{r}_{\mathrm{HD}}(t) = f(\vec{V}_{\mathrm{HD},a}(t)) \in \mathbb{R}^{N_{\mathrm{HD}}}$

- HR cell 被处理为 single-compartment neuron(单室神经元)
    - HR input current(HR 输入电流) 矢量 $\vec{I}_{\mathrm{HR}}(t)\in\mathbb{R}^{N_{\mathrm{HR}}}$
    - 动力学: $\begin{aligned}\tau_{s}\frac{\mathrm{d}\vec{I}_{\mathrm{HR}}}{\mathrm{d}t} = -\vec{I}_{\mathrm{HR}} + \mathbf{W}_{\mathrm{HD}\to\mathrm{HR}}\vec{r}_{\mathrm{HD}} + \vec{I}_{\mathrm{vel}\to\mathrm{HR}} + I_{\mathrm{inh},\mathrm{HR}}\vec{1}_{N_{\mathrm{HR}}}\end{aligned}$
        - $I_{\mathrm{inh},\mathrm{HR}}$: HR 集群的 global inhibition
    - HR firing rate(头旋转细胞放电率) 矢量 $\vec{r}_{\mathrm{HR}}(t) = f(\vec{I}_{\mathrm{HR}}(t)) \in \mathbb{R}^{N_{\mathrm{HR}}}$




# Local learning rule

> 对于 plastic synapse $j\to i$, postsynaptic neuron $i$ 必须是 HD cell. ($\mathbf{W}_{\mathrm{HD}\to\mathrm{HD}}$ 和 $\mathbf{W}_{\mathrm{HR}\to\mathrm{HD}}$)

- postsynaptic learning error(突触后学习误差): $E_{\mathrm{HD},i}(t) = f(V_{\mathrm{HD},a,i}(t)) - f(V_{\mathrm{HD},ss,i}(t)) \in \mathbb{R}$
    - $f(V_{\mathrm{HD},a,i}(t))$: visual teacher 诱导的真实输出
    - $f(V_{\mathrm{HD},ss,i}(t))$: distal input 单独预测出的输出

- presynaptic postsynaptic potential(突触前诱发的突触后电位) $P_{\mathrm{pre},j}(t)\in \mathbb{R}$
    - $j\to i$ 由 presynaptic neuron $j$ 产生的局部 post-synaptic potential(PSP)
    - 动力学: $\begin{aligned}\tau_{P}\frac{\mathrm{d}P_{\mathrm{pre},j}}{\mathrm{d}t} = -P_{\mathrm{pre},j} + r_{\mathrm{pre},j}\end{aligned}$
        - $r_{\mathrm{pre},j}$: presynaptic neuron 的 firing rate

- learning rule: $\begin{aligned}\Delta W_{j\to i} = \eta E_{\mathrm{HD},i}(t)P_{\mathrm{pre},j} = \eta[f(V_{\mathrm{HD},a,i}(t)) - f(p_{d\to a}V_{\mathrm{HD},d,i}(t))]P_{\mathrm{pre},j}(t)\end{aligned}$
    > 原文形式为 $\Delta W_{pre, post} = \eta [f(V_{a,post}) - f(pV_{d,post})]P_{pre}$
    - $\eta>0$: learning rate(学习率)
    - $f(\vec{V}_{\mathrm{HD},a})$ 可通过 backpropagation of axonal activity(轴突活动的反向传播) 到达树突远端区

- recurrent weight update
    - HD PSP 矢量 $\vec{P}_{\mathrm{HD}}(t)\in\mathbb{R}^{N_{\mathrm{HD}}}$; Error 矢量 $\vec{E}_{\mathrm{HD}}(t)\in\mathbb{R}^{N_{\mathrm{HD}}}$
    - 动力学: $\begin{aligned}\frac{\mathrm{d}\mathbf{W}_{\mathrm{HD}\to\mathrm{HD}}}{\mathrm{d}t} = \eta_{\mathrm{HD}\to\mathrm{HD}} \vec{E}_{\mathrm{HD}}(t)[\vec{P}_{\mathrm{HD}}(t)]^{\top}\end{aligned}$
- HR-to-HD weight update
    - HR PSP 矢量 $\vec{P}_{\mathrm{HR}}(t)\in\mathbb{R}^{N_{\mathrm{HR}}}$
    - 动力学: $\begin{aligned}\frac{\mathrm{d}\mathbf{W}_{\mathrm{HR}\to\mathrm{HD}}}{\mathrm{d}t} = \eta_{\mathrm{HR}\to\mathrm{HD}} \vec{E}_{\mathrm{HD}}(t)[\vec{P}_{\mathrm{HR}}(t)]^{\top}\end{aligned}$









# Decode head direction

- population vector average(群体矢量平均, PVA): $\begin{aligned}z_{\mathrm{PVA}}(t) = \frac{1}{N_{\mathrm{HD}}}\sum_{i=1}^{N_{\mathrm{HD}}}r_{\mathrm{HD},i}(t)e^{i\theta_{\mathrm{HD},i}}\end{aligned}$
    - 内部估计头朝向: $\begin{aligned}\hat{\theta}_{\mathrm{HD}}(t) = \arg\,z_{\mathrm{PVA}}(t)\end{aligned}$, 和 Clark 的 COM 其实一致

# Toy model 设计

- config
    - ring bins: $N_{\theta} = 48$
    - HD cells: $\vec{r}_{\mathrm{HD}} \in \mathbb{R}^{N_{\theta}}$
    - HR cells: $\vec{r}_{\mathrm{HR}}(t) = \begin{bmatrix} \vec{r}_{\mathrm{LHR}}(t) \\ \vec{r}_{\mathrm{RHR}}(t) \end{bmatrix} \in \mathbb{R}^{2N_{\theta}}$
        - left HR cells: $\vec{r}_{\mathrm{LHR}}(t) \in \mathbb{R}^{N_{\theta}}$
        - right HR cells: $\vec{r}_{\mathrm{RHR}}(t) \in \mathbb{R}^{N_{\theta}}$
    - plastic recurrent weight: $\mathbf{W}_{\mathrm{HD}\to\mathrm{HD}} \in \mathbb{R}^{N_{\theta}\times N_{\theta}}$
    - plastic HR-to-HD weight: $\mathbf{W}_{\mathrm{HR}\to\mathrm{HD}} = [\mathbf{W}_{\mathrm{LHR}\to\mathrm{HD}}, \mathbf{W}_{\mathrm{RHR}\to\mathrm{HD}}]\in \mathbb{R}^{N_{\theta}\times 2N_{\theta}}$
        - plastic left HR-to-HD weight: $\mathbf{W}_{\mathrm{LHR}\to\mathrm{HD}}\in \mathbb{R}^{N_{\theta}\times N_{\theta}}$
        - plastic right HR-to-HD weight: $\mathbf{W}_{\mathrm{RHR}\to\mathrm{HD}} \in \mathbb{R}^{N_{\theta}\times N_{\theta}}$
    - fixed HD-to-HR weight: $\mathbf{W}_{\mathrm{HD}\to\mathrm{HR}} = w_{\mathrm{HD}\to\mathrm{HR}} \begin{bmatrix} \mathbf{I}_{N_{\theta}} \\ \mathbf{I}_{N_{\theta}} \end{bmatrix} \in \mathbb{R}^{2N_{\theta}\times N_{\theta}}$

- data-stream
    - 外部变量更新: $\theta(t+\Delta t) = \theta(t) + v(t)\Delta t\quad (\mod{2\pi})$
        - 头朝向 $\theta(t)\in [0, 2\pi)$; 
        - (角)速度 $v(t)\in\mathbb{R}$ 通过 Ornstein-Uhlenbeck process(OU 过程) 生成: $\begin{aligned} v(t+\Delta t) = \left(1-\frac{\Delta t}{\tau_{v}}\right)v(t) + \sigma_{v}\sqrt{\Delta t}\xi(t)\end{aligned}$, 其中 $\xi(t)\sim \mathcal{N}(0,1)$
    - visual teacher: 
        1. toy model version(von Mises bump): $\begin{aligned}I_{\mathrm{vis}\to\mathrm{HD},i} = A_{\mathrm{vis}}\exp{[\kappa_{\mathrm{vis}}\cos{(\theta_{\mathrm{HD},i}-\theta(t))}]} - b_{\mathrm{vis}}\end{aligned}$
            - $\vec{I}_{\mathrm{vis}\to\mathrm{HD}}(t) \in \mathbb{R}^{N_{\theta}}$
        2. periodic Gaussian bump(*)
    - velocity input: $\vec{I}_{\mathrm{vel}\to\mathrm{HR}} = \begin{bmatrix} \vec{I}_{\mathrm{vel}\to\mathrm{LHR}}(t) \\ \vec{I}_{\mathrm{vel}\to\mathrm{RHR}}(t) \end{bmatrix}\in\mathbb{R}^{2N_{\theta}}$
        - $\vec{I}_{\mathrm{vel}\to\mathrm{LHR}}(t) = -k_{v}v(t)\vec{1}_{N_{\theta}}\in\mathbb{R}^{N_{\theta}}$;
        - $\vec{I}_{\mathrm{vel}\to\mathrm{RHR}}(t) = +k_{v}v(t)\vec{1}_{N_{\theta}}\in\mathbb{R}^{N_{\theta}}$;
    - HR update: $\begin{aligned}\tau_{s}\frac{\mathrm{d}\vec{I}_{\mathrm{HR}}}{\mathrm{d}t} = -\vec{I}_{\mathrm{HR}} + \mathbf{W}_{\mathrm{HD}\to\mathrm{HR}}\vec{r}_{\mathrm{HD}} + \vec{I}_{\mathrm{vel}\to\mathrm{HR}} - b_{\mathrm{HR}}\vec{1}_{2N_{\theta}}\end{aligned}$
        - $\vec{I}_{\mathrm{HR}}(t) \in \mathbb{R}^{2N_{\theta}}$
        - $\vec{r}_{\mathrm{HR}}(t) = f(\vec{I}_{\mathrm{HR}}(t)) \in \mathbb{R}^{2N_{\theta}}$
    - HD distal input update: $\begin{aligned}\tau_{s}\frac{\mathrm{d}\vec{I}_{\mathrm{HD},d}}{\mathrm{d}t} = -\vec{I}_{\mathrm{HD},d} + \mathbf{W}_{\mathrm{HD}\to\mathrm{HD}}\vec{r}_{\mathrm{HD}} + \mathbf{W}_{\mathrm{HR}\to\mathrm{HD}}\vec{r}_{\mathrm{HR}} - b_{\mathrm{HD}}\vec{1}_{N_{\theta}}\end{aligned}$
        - $\vec{I}_{\mathrm{HD},d}(t) \in \mathbb{R}^{N_{\theta}}$
    - assumption: 
        - $\vec{V}_{\mathrm{HD},d}(t) = \vec{I}_{\mathrm{HD},d}(t) \in \mathbb{R}^{N_{\theta}}$;
        - $\vec{V}_{\mathrm{HD},ss}(t) = p_{d\to a}\vec{V}_{\mathrm{HD},d}(t) \in \mathbb{R}^{N_{\theta}}$;
        - $\vec{V}_{\mathrm{HD},a}(t) = \vec{V}_{\mathrm{HD},ss}(t) + \vec{I}_{\mathrm{vis}\to\mathrm{HD}}(t)\in \mathbb{R}^{N_{\theta}}$;
        - $\vec{r}_{\mathrm{HD}}(t) = f(\vec{V}_{\mathrm{HD},a}(t)) \in \mathbb{R}^{N_{\theta}}$
    - learning error: $\vec{E}_{\mathrm{HD}}(t) = f(\vec{V}_{\mathrm{HD},a}(t)) - f(\vec{V}_{\mathrm{HD},ss}(t))\in\mathbb{R}^{N_{\theta}}$
    - PSP filters(突触后电压滤波):
        - $\begin{aligned}\tau_{P}\frac{\mathrm{d}\vec{P}_{\mathrm{HD}}}{\mathrm{d}t} = -\vec{P}_{\mathrm{HD}} + \vec{r}_{\mathrm{HD}},\quad \vec{P}_{\mathrm{HD}}(t)\in\mathbb{R}^{N_{\theta}}\end{aligned}$;
        - $\begin{aligned}\tau_{P}\frac{\mathrm{d}\vec{P}_{\mathrm{HR}}}{\mathrm{d}t} = -\vec{P}_{\mathrm{HR}} + \vec{r}_{\mathrm{HR}},\quad \vec{P}_{\mathrm{HR}}(t)\in\mathbb{R}^{2N_{\theta}}\end{aligned}$
    - weight update: $\lambda\geq 0$ 为 weight decay
        - $\begin{aligned}\frac{\mathrm{d}\mathbf{W}_{\mathrm{HD}\to\mathrm{HD}}}{\mathrm{d}t} = \eta_{\mathrm{HD}\to\mathrm{HD}}\vec{E}_{\mathrm{HD}}(\vec{P}_{\mathrm{HD}})^{\top} - \lambda \mathbf{W}_{\mathrm{HD}\to\mathrm{HD}}\end{aligned}$;
        - $\begin{aligned}\frac{\mathrm{d}\mathbf{W}_{\mathrm{HR}\to\mathrm{HD}}}{\mathrm{d}t} = -\eta_{\mathrm{HR}\to\mathrm{HD}}\vec{E}_{\mathrm{HD}}(\vec{P}_{\mathrm{HR}})^{\top} - \lambda \mathbf{W}_{\mathrm{HR}\to\mathrm{HD}}\end{aligned}$
        - 每次 weight update 后, 设定 $W_{\mathrm{HD}\to\mathrm{HD},ii}=0$ 避免自连接, 并且做 clipping $\mathbf{W} \leftarrow \mathrm{clip}{(\mathbf{W}, 0, \mathbf{W}_{\max})}$

- check
    1. 给定短暂 visual pulse 后令 $v(t)=0$, 检查解码位置 $\hat{\theta}_{\mathrm{HD}}$ 相比于 $\theta_{0}$ 是否稳定
    2. 检查 path integration: 给定已知 $v(t)$, 检查解码位置是否有 $\begin{aligned}\hat{\theta}_{\mathrm{HD}}(t) \approx \theta(0) + \int_{0}^{t}v(t^{\prime})\mathrm{d}t^{\prime}\end{aligned}$
    3. 检查 velocity gain: 用不同 constant $v$ 检查 bump velocity $\dot{\hat{\theta}}_{\mathrm{HD}}$, 拟合 $\dot{\hat{\theta}}_{\mathrm{HD}}\approx gv$, 理想结果是 $g\approx 1$
